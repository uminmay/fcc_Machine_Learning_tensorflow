#### Convolutional Layer

Different than Dense() layers, each CNN is made up of one or many convolutional layers. Goal is to find patterns within images
patterns -> say cat ears, nose, paws etc. Dense will learnin specific area(Global paterns), if we flip image it has to relearn. CNN scan entire image, determine features and send to Dense(). Convolution Layers output FEATURE MAP.. we will run many filters, and extract features.


- input size
- number of filters
- sample size of the filters

```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt
```

```python
(train_images, train_labels), (test_images, test_labels) = datasets.cifar10.load_data()
train_images, test_images = train_images/255.0, test_images/255.0

class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer', \
              'dog', 'frog', 'horse', 'ship', 'truck']
```

```python
IMG_INDEX = 506 #change index to check random images

plt.imshow(train_images[IMG_INDEX], cmap = plt.cm.binary)
plt.xlabel(class_names[train_labels[IMG_INDEX][0]])
plt.show()
```

## CNN Architecture

- Common CNN architecture -> stack of Conv2D and MaxPooling2D layers followed by a few densely connected layers.

```python
model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape = (32, 32, 3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
```

#### Layer 1
The input shape of our data will be 32, 32, 3 and we will process 32 filters of size 3x3 over input data. Also, relu.

#### Layer 2
Performs Max Pooling using 2x2 samples and stride of 2

#### Other layers

Similar but take input as feature map from previous layer, increased frequency of filters from 32 to 64, as our data shrinks , we can increase computational complexity.

```python
model.summary()
```

#### Add Dense layers

Forward from convolutional base

```python
model.add(layers.Flatten()) # take 4x4x64 to linear
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(10))
```

```python
model.summary()
```

### Training

```python
model.compile(optimizer = 'adam', \
             loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits = True), \
             metrics = ['accuracy'])
history = model.fit(train_images, train_labels, epochs = 10, \
                   validation_data = (test_images, test_labels))
```

### Evaluationg the model

We can determine how model performed by looking at its performance on the test data set.

```python
test_loss, test_acc = model.evaluate(test_images, test_labels, verbose = 2)
```

#### Fine Tuning | Using a Pretrained Model

```python
import os
import numpy as np
keras = tf.keras
```

#### Dataset

Cats_vs_Dogs dataset

```python
(raw_train, raw_validation, raw_test), metadata = tfds.load(
'cats_vs_dogs',
split=[
   tfds.Split.TRAIN.subsplit(tfds.percent[:80]),
   tfds.Split.TRAIN.subsplit(tfds.percent[80:90]),
   tfds.Split.TRAIN.subsplit(tfds.percent[90:])
],
with_info=True,
as_supervised=True,
)
```

```python
get_label_name = metadata.features['label'].int2str

for image,label in raw_train.take(2):
    plt.figure()
    plt.imshow(image)
    plt.title(get_label_name(label))
```

#### Data Preprocessing

#### Scaling all images to same dimentions

```python
IMG_SIZE = 160 #160x160

def format_example(image, label):
    image = tf.cast(image, tf.float32)
    image = (image/127.5) - 1
    image = tf.image.resize(image, (IMG_SIZE, IMG_SIZE))
    return image, label
```

Now we can apply this function to our images with map

```python
train = raw_train.map(format_example)
validation = raw_validation.map(format_example)
test = raw_test.map(format_example)
```

Lets take a look a images

```python
for image, label in train.take(2):
    plt.figure()
    plt.imshow(image)
    plt.title(get_label_name(label))
```

Finally we shuffle and batch the images

```python
BATCH_SIZE = 32
SHUFFLE_BUFFER_SIZE = 1000

train_batches = train.shuffle(SHUFFLE_BUFFER_SIZE).batch(BATCH_SIZE)
validation_batches = validation.batch(BATCH_SIZE)
test_bacthes = test.batch(BATCH_SIZE)
```

## Picking the PreTrained Model

MobileNet V2

1.4 million images and has 1000 different classes, we want to use the model but only convolutional base.

```python
IMG_SHAPE = ( IMG_SIZE, IMG_SIZE, 3)
base_model = tf.keras.applications.MobileNetV2(input_shape = IMG_SHAPE, \
                                               include_top = False, \
                                               weights = 'imagenet')
```

```python
base_model.summary()
```

```python
for image,_ in train_batches.take(1):
    pass
feature_batch = base_model(image)
print(feature_batch.shape)
```

#### Freezing the base

```python
base_model.trainable = False
base_model.summary()
```

#### Adding our classifier

Instead of flattening the feature map of the base layer we will use a global average pooling <br>
layer that will average the entire 5x5 area of each 2D feature map and return to use a single 1280 element vector per filter.

```python
global_average_layer = tf.keras.layers.GlobalAveragePooling2D()
```

```python
prediction_layer = keras.layers.Dense(1)
```

```python
model = tf.keras.Sequential([ \
                            base_model, \
                            global_average_layer, \
                            prediction_layer])
```

```python
model.summary()
```

### Training the model

```python
base_learning_rate = 0.0001
model.compile(optimizer = tf.keras.optimizers.RMSprop(learning_rate=base_learning_rate), \
             loss = tf.keras.losses.BinaryCrossentropy(from_logits = True), \
             metrics=['accuracy'])
```

```python
initial_epochs = 3
validation_steps = 20
loss0, accuracy0 = model.evaluate(validation_batches, steps = validation_steps)
```

```python
history = model.fit(train_batches, \
                   epochs = initial_epochs, \
                   validation_data = validation_batches)
acc = history.history['accuracy']
print(acc)
```
